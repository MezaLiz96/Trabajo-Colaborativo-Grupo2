```python
26#REGISTRO DE ALUMNOS CON PROGRAMACIÓN ESTRUCTURADA
# Lista de Alumnos
alumnos = []
```

```python
# Registro de Alumnos
def registrar_alumno():
    nombre = input("Ingrese nombre: ")
    edad = int(input("Ingrese edad: "))
    nota = float(input("Ingrese nota: "))

    alumno = {
        "nombre": nombre,
        "edad": edad,
        "nota": nota
    }

    alumnos.append(alumno)
    print("Alumno registrado correctamente\n")
```

```python
# Mostrar alumnos
def mostrar_alumnos():
    if len(alumnos) == 0:
        print("No hay alumnos registrados\n")
    else:
        for i, alumno in enumerate(alumnos):
            print(f"{i+1}. {alumno['nombre']} - Edad: {alumno['edad']} - Nota: {alumno['nota']}")
    print()
```

```python
# Promedio
def calcular_promedio():
    if len(alumnos) == 0:
        print("No hay datos para calcular promedio\n")
    else:
        suma = 0
        for alumno in alumnos:
            suma += alumno["nota"]

        promedio = suma / len(alumnos)
        print(f"Promedio de notas: {promedio}\n")
```
